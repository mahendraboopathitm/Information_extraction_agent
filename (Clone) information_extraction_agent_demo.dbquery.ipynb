```sql
WITH parsed_documents AS (
  SELECT
    path,
    ai_parse_document(
      content,
      map(
        'version', '2.0',
        'descriptionElementTypes', '*'
      )
    ) as parsed
  FROM
    read_files('/Volumes/data_engineering_workshop/agents/agentsdocs', format => 'binaryFile')
)
SELECT
  parsed AS content,
  ai_extract(
    parsed,
    '{
      "invoice_number" : {
        "type" : "string",
        "description" : "The unique identifier of the invoice"
      },
      "invoice_date" : {
        "type" : "string",
        "description" : "The date the invoice was issued in YYYY-MM-DD format"
      },
      "due_date" : {
        "type" : "string",
        "description" : "The date payment is due in YYYY-MM-DD format"
      },
      "po_number" : {
        "type" : "string",
        "description" : "The purchase order number associated with the invoice"
      },
      "payment_terms" : {
        "type" : "string",
        "description" : "The payment terms specified on the invoice"
      }
    }',
    options => map(
      'version', '2.0'
    )
  ) AS response
FROM parsed_documents
WHERE parsed IS NOT NULL
```
