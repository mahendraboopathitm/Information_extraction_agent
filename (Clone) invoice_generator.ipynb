```python
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from reportlab.lib.units import mm
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Spacer, HRFlowable
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.enums import TA_RIGHT, TA_CENTER
import os
 
def generate_random_invoice_data():
    """Generate a random invoice data dictionary."""
    supplier = random.choice(SUPPLIERS)
    buyer    = random.choice(BUYERS)
 
    # Random dates
    invoice_date = datetime.now() - timedelta(days=random.randint(0, 10))
    payment_term = random.choice(PAYMENT_TERMS)
    days_map     = {"Net 7 Days": 7, "Net 15 Days": 15, "Net 30 Days": 30,
                    "Net 45 Days": 45, "Immediate": 0}
    due_date     = invoice_date + timedelta(days=days_map.get(payment_term, 30))
 
    # Random line items (3 to 5)
    num_items = random.randint(3, 5)
    items     = random.sample(LINE_ITEMS_POOL, num_items)
    line_items = []
    for item in items:
        line_items.append({
            "description": item["desc"],
            "hsn_code":    item["hsn"],
            "qty":         random.randint(1, 200),
            "unit":        item["unit"],
            "unit_price":  item["price"] * random.uniform(0.9, 1.1),
            "tax_pct":     item["tax"]
        })
 
    cgst_rate, sgst_rate = random.choice(TAX_RATES)
    discount = round(random.randint(1000, 15000) / 500) * 500
 
    # Unique invoice number with timestamp
    inv_suffix = datetime.now().strftime("%Y%m%d%H%M%S")
    inv_number = f"INV-{inv_suffix}-{random.randint(100,999)}"
    po_number  = f"PO-SAP-{invoice_date.strftime('%Y%m%d')}-{random.randint(1,99):03d}"
 
    return {
        "invoice_number":   inv_number,
        "invoice_date":     invoice_date.strftime("%d-%b-%Y"),
        "due_date":         due_date.strftime("%d-%b-%Y"),
        "po_number":        po_number,
        "payment_terms":    payment_term,
        "supplier_name":    supplier["name"],
        "supplier_address": supplier["address"],
        "supplier_gstin":   supplier["gstin"],
        "supplier_pan":     supplier["pan"],
        "bill_to":          [buyer["name"]] + buyer["address"] + [f"GSTIN: {buyer['gstin']}"],
        "ship_to":          buyer["ship"],
        "line_items":       line_items,
        "cgst_rate":        cgst_rate,
        "sgst_rate":        sgst_rate,
        "discount":         discount,
        "bank_name":        supplier["bank"],
        "account_no":       supplier["acc"],
        "ifsc":             supplier["ifsc"],
        "branch":           supplier["branch"],
        "supplier_email":   supplier["email"],
        "supplier_phone":   supplier["phone"],
    }
 
 
def create_invoice_pdf(data, output_path):
    """Create a professional invoice PDF from data dict."""
    doc    = SimpleDocTemplate(output_path, pagesize=A4,
                               rightMargin=15*mm, leftMargin=15*mm,
                               topMargin=15*mm, bottomMargin=15*mm)
    styles = getSampleStyleSheet()
    story  = []
 
    # ── Styles ──
    co_style    = ParagraphStyle('Co', parent=styles['Normal'], fontSize=14,
                                 fontName='Helvetica-Bold', textColor=colors.HexColor('#1a3c5e'))
    inv_style   = ParagraphStyle('Inv', parent=styles['Normal'], fontSize=26,
                                 fontName='Helvetica-Bold', textColor=colors.HexColor('#1a3c5e'), alignment=TA_RIGHT)
    normal_s    = ParagraphStyle('N', parent=styles['Normal'], fontSize=9,
                                 textColor=colors.HexColor('#333333'), fontName='Helvetica')
    right_s     = ParagraphStyle('R', parent=styles['Normal'], fontSize=9,
                                 alignment=TA_RIGHT, fontName='Helvetica')
    big_tot_s   = ParagraphStyle('BT', parent=styles['Normal'], fontSize=13,
                                 fontName='Helvetica-Bold', textColor=colors.white, alignment=TA_RIGHT)
    footer_s    = ParagraphStyle('F', parent=styles['Normal'], fontSize=8,
                                 textColor=colors.HexColor('#888888'), alignment=TA_CENTER)
    item_s      = ParagraphStyle('Item', parent=styles['Normal'], fontSize=8,
                                 fontName='Helvetica', leading=10)
 
    # ── Header ──
    hdr = Table([[Paragraph(data['supplier_name'], co_style),
                  Paragraph("INVOICE", inv_style)]],
                colWidths=[95*mm, 85*mm])
    hdr.setStyle(TableStyle([('VALIGN',(0,0),(-1,-1),'TOP')]))
    story.append(hdr)
 
    addr = "<br/>".join(data['supplier_address']) + \
           f"<br/>GSTIN: {data['supplier_gstin']} | PAN: {data['supplier_pan']}"
    meta = (f"<b>Invoice No :</b>  {data['invoice_number']}<br/>"
            f"<b>Invoice Date :</b>  {data['invoice_date']}<br/>"
            f"<b>Due Date :</b>  {data['due_date']}<br/>"
            f"<b>PO Number :</b>  {data['po_number']}<br/>"
            f"<b>Payment Terms :</b>  {data['payment_terms']}")
    meta_t = Table([[Paragraph(addr, normal_s),
                     Paragraph(meta, ParagraphStyle('M', parent=styles['Normal'],
                               fontSize=9, alignment=TA_RIGHT, fontName='Helvetica'))]],
                   colWidths=[95*mm, 85*mm])
    meta_t.setStyle(TableStyle([('VALIGN',(0,0),(-1,-1),'TOP')]))
    story.append(meta_t)
    story.append(Spacer(1,5*mm))
    story.append(HRFlowable(width="100%", thickness=1.5, color=colors.HexColor('#1a3c5e')))
    story.append(Spacer(1,4*mm))
 
    # ── Bill To / Ship To ──
    bill = "<br/>".join(["<b>Bill To</b>"] + data['bill_to'])
    ship = "<br/>".join(["<b>Ship To</b>"] + data['ship_to'])
    bs   = Table([[Paragraph(bill, normal_s), Paragraph(ship, normal_s)]],
                 colWidths=[90*mm, 90*mm])
    bs.setStyle(TableStyle([
        ('BACKGROUND',(0,0),(-1,-1),colors.HexColor('#f0f4f8')),
        ('VALIGN',(0,0),(-1,-1),'TOP'),
        ('TOPPADDING',(0,0),(-1,-1),5), ('BOTTOMPADDING',(0,0),(-1,-1),5),
        ('LEFTPADDING',(0,0),(-1,-1),7), ('RIGHTPADDING',(0,0),(-1,-1),7),
    ]))
    story.append(bs)
    story.append(Spacer(1,5*mm))
 
    # ── Line Items ──
    col_h   = ['#', 'Item Description', 'HSN\nCode', 'Qty', 'Unit', 'Unit Price\n(INR)', 'Tax\n%', 'Amount\n(INR)']
    col_w   = [7*mm, 68*mm, 16*mm, 9*mm, 13*mm, 20*mm, 10*mm, 22*mm]
    t_data  = [col_h]
    subtotal = 0
    for i, item in enumerate(data['line_items'], 1):
        amount    = round(item['qty'] * item['unit_price'], 2)
        subtotal += amount
        t_data.append([
            str(i),
            Paragraph(item['description'], item_s),
            item['hsn_code'], str(item['qty']), item['unit'],
            f"{item['unit_price']:,.2f}", f"{item['tax_pct']}%", f"{amount:,.2f}"
        ])
    items_t = Table(t_data, colWidths=col_w, repeatRows=1)
    items_t.setStyle(TableStyle([
        ('BACKGROUND',(0,0),(-1,0),colors.HexColor('#1a3c5e')),
        ('TEXTCOLOR',(0,0),(-1,0),colors.white),
        ('FONTNAME',(0,0),(-1,0),'Helvetica-Bold'), ('FONTSIZE',(0,0),(-1,0),8),
        ('ALIGN',(0,0),(-1,0),'CENTER'),
        ('TOPPADDING',(0,0),(-1,0),5), ('BOTTOMPADDING',(0,0),(-1,0),5),
        ('FONTNAME',(0,1),(-1,-1),'Helvetica'), ('FONTSIZE',(0,1),(-1,-1),8),
        ('ALIGN',(3,1),(-1,-1),'CENTER'), ('ALIGN',(5,1),(-1,-1),'RIGHT'),
        ('ALIGN',(7,1),(-1,-1),'RIGHT'), ('VALIGN',(0,0),(-1,-1),'MIDDLE'),
        ('TOPPADDING',(0,1),(-1,-1),4), ('BOTTOMPADDING',(0,1),(-1,-1),4),
        ('LEFTPADDING',(0,0),(-1,-1),4), ('RIGHTPADDING',(0,0),(-1,-1),4),
        ('ROWBACKGROUNDS',(0,1),(-1,-1),[colors.white, colors.HexColor('#f7f9fb')]),
        ('GRID',(0,0),(-1,-1),0.4,colors.HexColor('#cccccc')),
        ('LINEBELOW',(0,0),(-1,0),1,colors.HexColor('#1a3c5e')),
    ]))
    story.append(items_t)
    story.append(Spacer(1,5*mm))
 
    # ── Totals ──
    cgst = round(subtotal * data['cgst_rate'] / 100, 2)
    sgst = round(subtotal * data['sgst_rate'] / 100, 2)
    net  = round(subtotal + cgst + sgst - data['discount'], 2)
 
    for label, val in [
        ("Subtotal",                        f"INR {subtotal:,.2f}"),
        (f"CGST ({data['cgst_rate']}%)",    f"INR {cgst:,.2f}"),
        (f"SGST ({data['sgst_rate']}%)",    f"INR {sgst:,.2f}"),
        ("Discount",                        f"- INR {data['discount']:,.2f}"),
    ]:
        t = Table([[Paragraph(label, normal_s), Paragraph(val, right_s)]],
                  colWidths=[130*mm, 50*mm])
        t.setStyle(TableStyle([('TOPPADDING',(0,0),(-1,-1),2),
                               ('BOTTOMPADDING',(0,0),(-1,-1),2)]))
        story.append(t)
 
    grand_t = Table([[Paragraph("Grand Total (Net Payable)", big_tot_s),
                      Paragraph(f"INR {net:,.2f}", big_tot_s)]],
                    colWidths=[130*mm, 50*mm])
    grand_t.setStyle(TableStyle([
        ('BACKGROUND',(0,0),(-1,-1),colors.HexColor('#1a3c5e')),
        ('TOPPADDING',(0,0),(-1,-1),7), ('BOTTOMPADDING',(0,0),(-1,-1),7),
        ('LEFTPADDING',(0,0),(0,0),8),  ('RIGHTPADDING',(1,0),(1,0),8),
    ]))
    story.append(grand_t)
    story.append(Spacer(1,5*mm))
 
    # ── Bank + Notes ──
    bank_txt  = (f"<b>Bank Details</b><br/>Bank Name : {data['bank_name']}<br/>"
                 f"Account No : {data['account_no']}<br/>"
                 f"IFSC Code : {data['ifsc']}<br/>Branch : {data['branch']}")
    notes_txt = ("<b>Notes / Terms</b><br/>Payment via NEFT/RTGS only.<br/>"
                 "Late payment attracts 1.5% interest/month.<br/>"
                 "Subject to jurisdiction of supplier's city.")
    bn = Table([[Paragraph(bank_txt, normal_s), Paragraph(notes_txt, normal_s)]],
               colWidths=[90*mm, 90*mm])
    bn.setStyle(TableStyle([
        ('BACKGROUND',(0,0),(-1,-1),colors.HexColor('#f0f4f8')),
        ('VALIGN',(0,0),(-1,-1),'TOP'),
        ('TOPPADDING',(0,0),(-1,-1),6), ('BOTTOMPADDING',(0,0),(-1,-1),6),
        ('LEFTPADDING',(0,0),(-1,-1),7), ('RIGHTPADDING',(0,0),(-1,-1),7),
        ('LINEAFTER',(0,0),(0,-1),0.5,colors.HexColor('#cccccc')),
    ]))
    story.append(bn)
    story.append(Spacer(1,5*mm))
    story.append(HRFlowable(width="100%", thickness=0.8, color=colors.HexColor('#cccccc')))
    story.append(Spacer(1,2*mm))
    story.append(Paragraph(
        f"Computer-generated invoice | {data['supplier_name']} | "
        f"{data['supplier_email']} | {data['supplier_phone']}",
        footer_s
    ))
 
    doc.build(story)
    return net  # return grand total for logging
 
 
print("✅ Invoice generator function ready!")
 
```

```python
invoice_data = generate_random_invoice_data()
 
# Create unique filename using invoice number
safe_inv_num = invoice_data['invoice_number'].replace('/', '-')
filename     = f"invoice_{safe_inv_num}.pdf"
output_path  = f"{VOLUME_PATH}/{filename}"
 
# Generate the PDF
grand_total = create_invoice_pdf(invoice_data, output_path)
 
print(f"✅ Invoice PDF generated and saved!")
print(f"   📄 File         : {filename}")
print(f"   🏢 Supplier     : {invoice_data['supplier_name']}")
print(f"   🛒 Buyer        : {invoice_data['bill_to'][0]}")
print(f"   📅 Invoice Date : {invoice_data['invoice_date']}")
print(f"   📅 Due Date     : {invoice_data['due_date']}")
print(f"   💰 Grand Total  : INR {grand_total:,.2f}")
print(f"   📦 Saved to     : {output_path}")
```
